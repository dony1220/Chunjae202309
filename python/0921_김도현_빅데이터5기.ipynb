```python
#상속
#super class
class Tree:
    kingdom = 'ang'
    leaf = True
    fruit = 'fruit_name'
    
    def flower(self):
        print("꼬치 피었습니다.")
```

```python
class FruitTree(Tree):
    def __init__(self, fruit):
        self.fruit = fruit
    def observe_fruit(self):
        print(f'나무에 달린열매는 {self.fruit}입니다.')
```

```python
FruitTree
```

```python
oragne_tree = FruitTree("orange")
oragne_tree
```

```python
oragne_tree.observe_fruit()
```
