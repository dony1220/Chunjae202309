# 해당 프로세스들을 클래스로 구현

```python
#타이타닉 csv 파일 불러오기
import numpy as np
import pandas as pd
df = pd.read_csv(r"C:\Users\user\dohyeon/titanic.csv", encoding='cp949')
df
```

```python
import pandas as pd

class Titanic:
    def __init__(self, data_file):
        self.data = pd.read_csv(data_file, encoding='cp949')
#클래스초기화

    def preprocess_data(self):
        # Sex 칼럼 삭제
        self.data.drop(columns=['Sex'], inplace=True)

        # 이름을 기반으로 성별을 파악하여 Sex column에 등록
        self.data['Sex'] = self.data['Name'].apply(self.gender)

        # Age 항목의 결측치 처리
        self.missing_age()

        # 남성이며 나이가 10대인 인원을 'check' column에 저장
        self.data['Check column'] = self.data.apply(self.check_mage10_gender, axis=1)
    def missing_age(self):
        # Age 항목의 결측치를 0으로 대체
        self.data['Age'].fillna(0, inplace=True)
    def check_mage10_gender(self, row):
        if row['Sex'] == 'Male' and 10 <= row['Age'] <= 19:
            return 'O'
        else:
            return 'X'
    #그냥 gender를 한 번에 연산이 가능하게 할 순 없나
    def gender(self, name):
        # 이름에서 "Mr." 또는 "Master."가 들어가면 남성으로, "Mrs." 또는 "Miss."가 들어가면 여성으로 분류
        #이거 말고는 여성남성 분류가 힘들어보임, 결측치가 있다면 어쩔수없음. 
        if "Mr." in name or "Master." in name:
            return 'Male'
        elif "Mrs." in name or "Miss." in name:
            return 'Female'
        else:
            return None
        
    def output_data(self, ouput_f):
        self.data.to_csv(ouput_f, sep = '\t')
        
    def csv_data(self, csv_f):
        self.data.to_csv(csv_f)
                      
input_f = "C:/Users/user/dohyeon/titanic.csv"
ouput_f = 'Result.tsv'
csv_f = 'ct.csv'

processor = Titanic(input_f)
processor.preprocess_data()
processor.csv_data(csv_f)
```
