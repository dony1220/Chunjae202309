# 문제 16 알고리즘, 범위 안의 숫자 x를 찾아내는 알고리즘
## 출력값 예시 : 몇 번째 위치에 있는지, 몇 차례만에 찾아냈는지.

```python
def binary_search(arr, x):
    left = 0
    right = len(arr) - 1
    count = 0  # 몇 번 비교했는지 카운트

    while left <= right:
        count += 1 # 비교카운트
        mid = (left + right) // 2  # 중간값 계산
        mid_value = arr[mid]

        if mid_value == x:
            return mid, count  # 찾았을 때 위치와 비교 횟수 반환
        elif mid_value < x:
            left = mid + 1
        else:
            right = mid - 1

    return -1, count  # 찾지 못했을 때 -1과 비교 횟수 반환

arr=[]
for i in range(1,101):
    arr.append(i)
x = int(input())  # 찾으려는 숫자

result, comparisons = binary_search(arr, x)
if result != -1:
    print(f"{x}는 배열의 {result}번째 위치에 있습니다.")
    print(f"{comparisons} 번의 비교를 통해 찾았습니다.")
else:
    print(f"{x}는 배열 안에 없습니다.")

```
